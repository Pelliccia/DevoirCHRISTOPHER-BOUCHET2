```python
import numpy as numpy
import math as math
def darcysolver(end):
    
    #les conditions initiales
    c_i = numpy.load("init_cond_c.npy")
    t=10**(-5)
    Ra=500
    k=0.1
    k_2=0.01
    h=10**(-8)
    delta=5/100
    delta_2=25/10000
    c_t=c_i
    while t <= end:
        
        #Calculons la matrice L
        
        a=-2*Ra/(t*delta_2)-k_2
        b=Ra/(t*delta_2)
    
        l=99
        L = [[0 for x in range(0,l)] for y in range(0,l)] 
        for i in range (0,99) : 
            for j in range (0,99) : 
                if i==j : 
                    L[i][j]=a
                if j==i+1 :
                    L[i][j]=b
                if j==i-1 :
                    L[i][j]=b
               
        #calculons L^-1
        inv=numpy.linalg.inv(L)
            
        #calculons W    
        W = [0 for x in range(l)] 
        
        for i in range(len(c_t)):
            for j in range(len(c_t)):
                W[i]=W[i]+ inv[i][j] *(-0.01)* c_t[j]

        #la matrice M
    
        c=(-1)/(2*delta_2)+(k_2*t)/Ra
        
        
        M = [[0 for x in range(l)] for y in range(l)] 
        for i in range (0,99) : 
            for j in range (0,99) :
                m=j
                d=(1/4)*(delta_2+m/2)
                if i==j : 
                    M[i][j]=c
                if j==i+1 :
                    M[i][j]=d
                if j==i-1 :
                    M[i][j]=d
        
        #calculons dc/dt (nous noterons dc/dt = derivative)
        derivative = [0 for x in range(0,l)]
        
        #pour cela nous devons calculer Mc 
        Mc=[0 for x in range(0,l)]  
        
        for i in range(len(c_t)):
            for j in range(len(c_t)):
                Mc[i]=Mc[i]+(1/t)*M[i][j] * c_t[j]
        
        for j in range (0,99) :    
            derivative[j]=((Ra/(math.pi)/t)**(0.5))*W[j]*math.exp(-(j*delta)*(j*delta))+Mc[j]
        
        k1=[0 for x in range(0,l)]        
    
        for j in range (0,l) :
            k1[j]=h*derivative[j]
        
        #calculons k2, pour cela il faut calculer M et L à l'instant t+h/2
        
        k2=[0 for x in range(0,99)]
        
        
        for j in range (0,99) :
            c_t[j]=c_t[j]+k1[j]/2
        
        
        t=t+h/2
        
        #copie des calculs ci-dessus
        
        #Calculons la matrice L
        
        a=-2*Ra/(t*delta_2)-k_2
        b=Ra/(t*delta_2)
    
        l=99
        L = [[0 for x in range(0,l)] for y in range(0,l)] 
        for i in range (0,99) : 
            for j in range (0,99) : 
                if i==j : 
                    L[i][j]=a
                if j==i+1 :
                    L[i][j]=b
                if j==i-1 :
                    L[i][j]=b
               
        #calculons L^-1
        inv=numpy.linalg.inv(L)
            
        #calculons W    
        W = [0 for x in range(l)] 
        
        for i in range(len(c_t)):
            for j in range(len(c_t)):
                W[i]=W[i]+inv[i][j] *(-0.01)* c_t[j]

        #la matrice M
    
        c=(-1)/(2*delta_2)+(k_2*t)/Ra
        
        
        M = [[0 for x in range(l)] for y in range(l)] 
        for i in range (0,99) : 
            for j in range (0,99) :
                m=j
                d=(1/4)*(delta_2+m/2)
                if i==j : 
                    M[i][j]=c
                if j==i+1 :
                    M[i][j]=d
                if j==i-1 :
                    M[i][j]=d
        
        #calculons dc/dt (nous noterons dc/dt = derivative)
        derivative = [0 for x in range(0,99)]
        
        #pour cela nous devons calculer Mc 
        Mc=[0 for x in range(0,l)]  
        for i in range(len(c_t)):
            for j in range(len(c_t)):
                Mc[i]=Mc[i]+(1/t)* M[i][j] * c_t[j]
        
        for j in range (0,99) :    
            derivative[j]=((Ra/(math.pi)*t)**0.5)*W[j]*math.exp(-(j*delta)*(j*delta))+Mc[j]    
        
        
        
        for j in range (0,l) : 
            k2[j]=h*derivative[j]
        
        #calculons k3
        k3=[0 for x in range(0,l)]
        
        c_t[j]=c_t[j]-k1[j]/2+k2[j]/2
        
        
        #copier coller
        
        #Calculons la matrice L
        
        a=-2*Ra/(t*delta_2)-k_2
        b=Ra/(t*delta_2)
    
        l=99
        L = [[0 for x in range(0,l)] for y in range(0,l)] 
        for i in range (0,99) : 
            for j in range (0,99) : 
                if i==j : 
                    L[i][j]=a
                if j==i+1 :
                    L[i][j]=b
                if j==i-1 :
                    L[i][j]=b
               
        #calculons L^-1
        inv=numpy.linalg.inv(L)
            
        #calculons W    
        W = [0 for x in range(l)] 
        
        for i in range(len(c_t)):
            for j in range(len(c_t)):
                W[i]=W[i] + inv[i][j] *(-0.01)* c_t[j]

        #la matrice M
    
        c=(-1)/(2*delta_2)+(k_2*t)/Ra
        
        
        M = [[0 for x in range(l)] for y in range(l)] 
        for i in range (0,99) : 
            for j in range (0,99) :
                m=j
                d=(1/4)*(delta_2+m/2)
                if i==j : 
                    M[i][j]=c
                if j==i+1 :
                    M[i][j]=d
                if j==i-1 :
                    M[i][j]=d
        
        #calculons dc/dt (nous noterons dc/dt = derivative)
        derivative = [0 for x in range(l)]
        
        #pour cela nous devons calculer Mc 
        Mc=[0 for x in range(0,l)]  
        for i in range(len(c_t)):
            for j in range(len(L)):
                Mc[i]=Mc[i]+ (1/t)* M[i][j] * c_t[j]
        
        for j in range (0,99) :    
            derivative[j]=((Ra/(math.pi)*t)**0.5)*W[j]*math.exp(-(j*delta)*(j*delta))+Mc[j]
                   
        for j in range (0,l) : 
            k3[j]=h*derivative[j]
                
       
        #calculons k4
        t=t+h/2
        for j in range (0,l) :
            c_t[j]=c_t[j]-k2[j]/2+k3[j]
        
        
        #copier coller
        
        
        #Calculons la matrice L
        
        a=-2*Ra/(t*delta_2)-k_2
        b=Ra/(t*delta_2)
    
        l=99
        L = [[0 for x in range(0,l)] for y in range(0,l)] 
        for i in range (0,99) : 
            for j in range (0,99) : 
                if i==j : 
                    L[i][j]=a
                if j==i+1 :
                    L[i][j]=b
                if j==i-1 :
                    L[i][j]=b
               
        #calculons L^-1
        inv=numpy.linalg.inv(L)
            
        #calculons W    
        W = [0 for x in range(l)] 
        
        for i in range(len(c_t)):
            for j in range(len(c_t)):
                W[i]=W[i]+ inv[i][j] *(-0.01)* c_t[j]

        #la matrice M
    
        c=(-1)/(2*delta_2)+(k_2*t)/Ra
        
        
        M = [[0 for x in range(l)] for y in range(l)] 
        for i in range (0,99) : 
            for j in range (0,99) :
                m=j
                d=(1/4)*(delta_2+m/2)
                if i==j : 
                    M[i][j]=c
                if j==i+1 :
                    M[i][j]=d
                if j==i-1 :
                    M[i][j]=d
        
        #calculons dc/dt (nous noterons dc/dt = derivative)
        derivative = [0 for x in range(l)]
        
        #pour cela nous devons calculer Mc 
        Mc=[0 for x in range(0,l)]  
        for i in range(len(c_t)):
            for j in range(len(L)):
                Mc[i]=Mc[i]+(1/t)* M[i][j] * c_t[j]
        
        for j in range (0,99) :    
            derivative[j]=((Ra/(math.pi)*t)**0.5)*W[j]*math.exp(-(j*delta)*(j*delta))+Mc[j]
        
               
        
        
        k4=[0 for x in range(0,l)]
        for j in range (0,l): 
            k4[j]=h*derivative[j]
    
        
        for j in range (0,l):
            c_t[j]=c_t[j]-k3[j]+(k1[j]+2*k2[j]+2*k3[j]+k4[j])/6
        print (c_t)
```
